## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
'''My lane is a ranking/scoring problem.

The goal is not simply to predict whether content is declining or not. The real goal is to help prioritize which content should be reviewed first.

The model output would be a priority score that ranks content items from highest urgency to lowest urgency for refresh or review.'''
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
''' The target is the likelihood that a content item is declining in performance.

For the starter dataset, a practical proxy target is:

is_declining_label

However, this label is derived from trend_direction, which itself comes from trend_pct.

Because of this, trend_direction and trend_pct must never be used as features.

The target is observed in the dataset and represents an outcome rather than a manually written rule.'''
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
''' Since this is a ranking task, the primary success metric is:

Precision@K

The question is:

Of the top K pages that the model recommends for refresh, how many actually needed attention?

A secondary metric could be recall, but Precision@K better matches the business decision because editor time is limited. '''
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
'''The decision being improved is:

Which pages should an editor refresh first?

This is not a prediction problem for its own sake.

The output helps prioritize work so that editors spend time on pages that matter most.'''
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
'''A simple rule could be:

Refresh every page with traffic below a fixed threshold.

However, content performance depends on many interacting signals:

CTR
Engagement rate
Scroll rate
Position changes
Content type
Historical performance

These relationships are messy and change over time.

Machine learning can combine many weak signals into a better prioritization score than a manually written rule.'''
```
